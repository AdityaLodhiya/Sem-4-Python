```python
df=pd.read_csv("auto-mpg.csv")
```

```python
df['horsepower']=df['horsepower'].replace('?',df[df['horsepower']!='?']["horsepower"].astype("int64").mean())
df['horsepower']=df['horsepower'].astype('int64')
print(df['horsepower'].unique())
```
