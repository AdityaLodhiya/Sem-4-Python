```python
import networkx as nx
```

```python
G=nx.Graph()
G.add_node("A")
G.add_node("B")
G.add_node("C")
G.add_edge("A","B")
G.add_edge("B","C")
G.add_edge("C","A")
G.add_edge("A","A")
nx.draw(G,with_labels=True)
```

```python
G=nx.DiGraph()
G.add_node("A")
G.add_node("B")
G.add_node("C")
G.add_edge("A","B")
G.add_edge("B","C")
G.add_edge("C","A")
G.add_edge("A","A")
nx.draw(G,with_labels=True)
```

```python
G=nx.Graph()
G.add_nodes_from([1,2,3])
G.add_edges_from([(1,2),(2,3),(3,1),(1,1)])
nx.draw(G,with_labels=True)
```
