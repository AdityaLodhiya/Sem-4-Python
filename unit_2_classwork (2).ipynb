# 1.You are given marks of 12 students in 5 subjects: Math, Physics, Chemistry, Biology, and English.

1. Create a DataFrame df_scores with random integer values between 40 and 100.

2. Analyze the relationship between all subjects.

3. Visualize the relationships using a heatmap such that:

The color scale is centered properly for negative to positive relationships

Strong positive and negative relationships are clearly distinguishable using contrasting colors

Each cell is clearly separated by visible borders

The exact numeric relationship values are displayed inside each cell

A legend for colors is visible

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

np.random.seed(42)

df_scores = pd.DataFrame({
    "Math": np.random.randint(40, 100, 12),
    "Physics": np.random.randint(40, 100, 12),
    "Chemistry": np.random.randint(40, 100, 12),
    "Biology": np.random.randint(40, 100, 12),
    "English": np.random.randint(40, 100, 12)
})

print(df_scores)
```

```python
print(df_scores.corr())
print(sns.heatmap(df_scores.corr(),vmin=-1,vmax=1,linecolor="red",linewidths=2,annot=True))
```
